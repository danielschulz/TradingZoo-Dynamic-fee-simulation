```python
import matplotlib.pyplot as plt
import numpy as np
import torch
%matplotlib inline
```

```python
scores = torch.load('scores.pth')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))
ax.grid(True)
# ax.set_ylim(-0.6, 2)

ax.plot(scores);
```

```python
window_size = 200
running_mean_scores = [np.mean(scores[max(i-window_size+1, 0):i+1]) for i in range(len(scores))]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))
ax.grid(True)
ax.plot(running_mean_scores, color='orange');
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))
ax.grid(True)
ax.plot(scores, alpha=0.8)
ax.plot(running_mean_scores, color='orange', alpha=0.8)
ax.set_ylim(-0.2, 1);
```
